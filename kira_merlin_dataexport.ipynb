```python
Cab = df[df['Location']=='Cab']
Cargo = df[df['Location'] == 'Cargo']
Out = df[df['Location'] == 'Out']
Outside = df[df['Location'] == 'Outside']
```

```python
Cab['Temperature'].plot()
```

```python
Locations = [Cab, Cargo, Out, Outside]
LocName = ["Cab", "Cargo", 'Out', 'Outside']
```

```python
for i in range(4):
    Locations[i]['Temperature'].plot()
    Locations[i]['Wet Bulb Temp'].plot()
    Locations[i]['Globe Temperature'].plot()
    plt.title(LocName[i])
    plt.figure()
    
```

```python
MerlinCargo = df[(df['Technology'] == 'Merlin') & (df['Location'] == 'Cargo')]
MerlinCargoDates = MerlinCargo[(MerlinCargo['Date'] == '6/27/24')]
BaselineCargo = df[(df['Technology'] == 'Baseline') & (df['Location'] == 'Cargo')]
BaselineCargoDates = BaselineCargo[(BaselineCargo['Date'] == '6/27/24')]
```

```python
MerlinCargoDates['Temperature'].plot()
# MerlinCargoDates['Wet Bulb Temp'].plot()
MerlinCargoDates['Globe Temperature'].plot()
MerlinCargoDates['Heat Index'].plot()
plt.title('Merlin')
plt.legend()
plt.figure()

BaselineCargoDates['Temperature'].plot()
# BaselineCargoDates['Wet Bulb Temp'].plot()
BaselineCargoDates['Globe Temperature'].plot()
BaselineCargoDates['Heat Index'].plot()
plt.title('Baseline')
plt.legend()
plt.figure()
```

```python
MerlinCargoDates['Heat Index'].plot()
BaselineCargoDates['Heat Index'].plot()
plt.title('Merlin vs Baseline')
plt.legend(["Merlin", "Baseline"])
plt.xlabel("Date Time")
plt.ylabel("Degrees (C)")
```
